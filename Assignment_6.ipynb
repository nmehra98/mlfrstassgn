```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

```
df=pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv")
df
```

```
plt.plot([1,2,3,4],linewidth=5.0)     #It will take points as y axis with intervals of 0.5
plt.ylabel("Y axis")
plt.xlabel("X axis")
plt.show()
```

```
plt.plot(df["MaxTemp"],df["MinTemp"])
plt.ylabel("Min Temp")
plt.xlabel("Max Temp")
plt.show()
```

```
plt.plot([1,2,3,4],[5,6,7,8],'rs')    
plt.ylabel("Y axis")
plt.xlabel("X axis")
plt.show()
```

```
plt.scatter(df["MaxTemp"],df["MinTemp"],marker="^",c="g")
plt.ylabel("Min Temp")
plt.xlabel("Max Temp")
plt.show()
```

```
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.figure(1)
plt.subplot(211)   #numrows,numcols,fignum
plt.plot(t1,t1,"bo",label="t1")
plt.plot(t2,t2**2,"g",label="t2")
plt.title("Subplots")
plt.legend

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),"r--")
plt.grid(True)
plt.text(2.18,0.8,"r*cos func")
plt.show()
```
