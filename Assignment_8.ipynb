```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.preprocessing import*
%matplotlib inline
```

```
df=pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/house-price-prediction/master/kc_house_data.csv")
df
```

```
df.dtypes
```

```
del df["date"]
```

```
a = (df-df.min())/(df.max()-df.min())
a
```

```
a1 = (a-a.mean())/(a.max()-a.min())
a1
```

```
a2 = (a1-a1.median())/(a1.max()-a1.min())
a2
```

```
a2.std()
```

```
z=np.power(a2["bedrooms"],(1/3))
z
```

```
plt.figure(figsize=(10,8))
corr=df.corr()
sns.heatmap(corr,annot=True)
```

```
df.plot(x="price",y="sqft_living")
plt.xlabel("Price")
plt.ylabel("sqft_living")
plt.show()
```

```
x=np.array(df["price"]).reshape(21613,1)
x
```

```
y=np.array(df["sqft_living"]).reshape(21613,1)
y
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(y_train,x_train)
```

```
regressor.coef_
```

```
y_pred=regressor.predict(x_test)
y_pred
```

```
df=pd.DataFrame({"Actual":[x_test],"predicted":[y_pred]})
df
```
