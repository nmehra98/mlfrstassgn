```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.preprocessing import*
%matplotlib inline
```

```
from sklearn.datasets import load_digits

digit=load_digits()
x=digit.data
y=digit.target

x.shape,y.shape
```

```
from sklearn.preprocessing import StandardScaler
x_std = StandardScaler().fit_transform(x)
x_std
```

```
from sklearn.decomposition import PCA
pca = PCA(n_components=8)
pca.fit(x_std)
```

```
principalComponents = pca.fit_transform(x_std)
principalComponents
```

```
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2','principal component 3','principal component 4','principal component 5','principal component 6','principal component 7','principal component 8'])
principalDf

# new x
```

```
pca.explained_variance_ratio_
```

```
from sklearn.model_selection import train_test_split
principalDf_train,principalDf_test,y_train,y_test=train_test_split(principalDf,y,test_size=0.3)
```

```
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
reg.fit(principalDf_train,y_train)
```

```
pred = reg.predict(principalDf_test)
```

```
from sklearn import metrics
metrics.accuracy_score(y_test,pred)
```

```
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)
```
